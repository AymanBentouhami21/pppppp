**encode data**

```python
def encode_form(data):
    form_column = data['form']
    form_column = form_column.apply(lambda x: list(x))
    
    mapping_dict = {'L':-1,'D':0,'W':1}
    form_column = form_column.apply(lambda x: [mapping_dict[char] for char in x])
    form_column = form_column.apply(lambda x: [val*weight for val,weight in zip(x,[0.6,0.7,0.8,0.9,1.0])])
    form_column = form_column.apply(lambda x: sum(x))
    data['form_index'] = form_column
    return data
standings_current_df = encode_form(standings_current_df)
standings_current_df.drop(columns=['form','status','goalsDiff','points','home_goals_for','home_goals_against','away_goals_for','away_goals_against'],inplace=True)
standings_current_df.head(5)
```

## get historical h2h data for fixture

```python
season_v3_path = r'C:\Users\admin\Documents\Personal Work\Projects\Footbal Project\TRY 2\data\seasons\seasons_v3'
file_list = os.listdir(season_v3_path)
file_list.pop()
```

```python
def get_folder_data(file_list,folder_path):
    data_list = []
    for file in file_list:
        file_path = os.path.join(folder_path,file)
        with open(file_path,'r') as file:
            data = json.load(file)
            data_list.append(data)
    #we will flattent the list
    data_list = [item for sublist in data_list for item in sublist ]
    return data_list
#pd.DataFrame(get_folder_data(file_list,season_v3_path))
```

```python
all_fixture_df = pd.DataFrame(get_folder_data(file_list,season_v3_path))
all_fixture_df.head(5)
```

```python
all_fixture_df['fixture_name'] = all_fixture_df.apply(lambda x: f"{x['home_team_name']} vs {x['away_team_name']}" if x['home_team_name'] < x['away_team_name'] else f"{x['away_team_name']} vs {x['home_team_name']}",axis=1)
all_fixture_df.head(5)
```

```python
all_fixture_df['home_team_outcome'] = all_fixture_df.apply(lambda x: 'W' if x['score_fulltime_home'] > x['score_fulltime_away'] else ('D' if x['score_fulltime_home'] == x['score_fulltime_away'] else 'L'),axis=1)
all_fixture_df['home'] = all_fixture_df.apply(lambda x: 1 if x['home_team_name'] < x['away_team_name'] else 0,axis=1)
all_fixture_df.head(5)
```

```python
all_fixture_df['date'].isnull().sum()
```

```python
all_fixture_df['date'] = pd.to_datetime(all_fixture_df['date'])
all_fixture_df['date'] = all_fixture_df['date'].dt.tz_localize(None) ## cant compare tz aware with tz naive 
```

```python
all_fixture_df[['date','home_team_outcome']].isnull().sum()
```

```python
def get_weighted_outcome(data):
    mapping_dict = {'W':1,'D':0,'L':-1}
    data['encoded_home_team_outcome'] = data.apply(lambda x: mapping_dict[x['home_team_outcome']],axis=1)
    #weight should be by season, premier leage starts at the 8 season and ends in the 5th ,except for 2020(covid), so is the 2019 season wich ended in the 7 month becasue matches of the 4 and 5 th month werent played
    start_date = pd.to_datetime('2010-08-01')
    start_dates_list = [start_date + pd.DateOffset(years=i) for i in range(14)]
    end_date = pd.to_datetime('2011-07-30')
    end_dates_list = [end_date + pd.DateOffset(years=i) for i in range(14)]
    date_ranges = [tuple([date_minor,date_major,season]) for date_minor,date_major,season in zip(start_dates_list,end_dates_list,range(2011,2025))] #iff we start with 2010 the 2010 season will be weighted with 0

    
    season_column = []
    for _ , row in data.iterrows():
        for date_range in date_ranges:
            if row['date'] > date_range[0] and row['date'] < date_range[1]:
                season_column.append(date_range[2] - 2010)
    data['season'] = season_column
    data['weighted_outcome'] = data['season'] * data['encoded_home_team_outcome']
    return data
all_fixture_df = get_weighted_outcome(all_fixture_df)
all_fixture_df.drop(columns=['home_team_name','away_team_name','home_team_winner','away_team_winner','away_team_id','home_team_id','score_fulltime_home','score_fulltime_away','date'],inplace=True)
all_fixture_df.head(5)
```

```python
all_fixture_df = all_fixture_df.groupby(['fixture_name','home'])[['weighted_outcome']].sum().reset_index()
all_fixture_df.head(5)
```

```python
merged = pd.merge(merged,all_fixture_df,left_on=['fixture_name','home_a'],right_on=['fixture_name','home'])
merged.drop(columns='home',inplace=True)
merged
```
